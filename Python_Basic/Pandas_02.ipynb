```python
def rand_func(x):
    return x ** 2
rand_func(5)
```

```python
def plus_one(x):
    return x + 1
```

```python
df_new['new_hair'] = df_new['hair'].apply(plus_one)
df_new
```

```python
pivot_df = pd.pivot_table(df_new, index = 'type', aggfunc = np.sum)
```
